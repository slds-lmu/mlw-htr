```python
import numpy as np
import pandas as pd
from torchmetrics import CharErrorRate
```

```python
df = pd.read_json('../../data/processed/test_predictions.json')
```

```python
metric = CharErrorRate()
```

```python
df.head()
```

```python
counter = 0
for row in df.iterrows():
    lemma = row[1][1]
    prediction = row[1][2]
    if lemma != prediction:
        print(lemma, prediction)
        counter += 1
```

```python
counter
```

```python
df['CER'] = list(map(lambda e: metric(preds=e[1][2], target=e[1][1]).item(), df.iterrows()))
df['Correct'] = list(map(lambda e: 1 if e[1][2] == e[1][1] else 0, df.iterrows()))
```

```python
df = df.rename(columns={"lemma": "Label", "prediction": "Prediction"})
```

```python
df['Model'] = ['SWIN+GPT2'] * len(df)
```

```python
df = df.drop(columns=['id'])
```

```python
df.to_json('results_swin_gpt2.json')
```

```python
df.to_excel('results_swin_gpt2.xlsx', index=False)
```
